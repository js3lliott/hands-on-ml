Now that we've read in our data and separated it into our features and labels, let's check out one of the images in the dataset.

```python
some_digit = X[0]
some_digit_image = some_digit.reshape(28, 28)

plt.imshow(some_digit_image, cmap=mpl.cm.binary, interpolation="nearest")
plt.axis("off")
plt.show();
```

These next few functions are provided by Aurelien Geron from his book Hands-On Machine Learning with Scikit-Learn and Tensorflow: Concepts, Tools, and Techniques to Build Intelligent Systems.

```python
def plot_digit(data):
    image = data.reshape(28, 28)
    plt.imshow(image, cmap = mpl.cm.binary,
               interpolation="nearest")
    plt.axis("off")
```

```python
def plot_digits(instances, images_per_row=10, **options):
    size = 28
    images_per_row = min(len(instances), images_per_row)
    images = [instance.reshape(size,size) for instance in instances]
    n_rows = (len(instances) - 1) // images_per_row + 1
    row_images = []
    n_empty = n_rows * images_per_row - len(instances)
    images.append(np.zeros((size, size * n_empty)))
    for row in range(n_rows):
        rimages = images[row * images_per_row : (row + 1) * images_per_row]
        row_images.append(np.concatenate(rimages, axis=1))
    image = np.concatenate(row_images, axis=0)
    plt.imshow(image, cmap = mpl.cm.binary, **options)
    plt.axis("off")
```

```python
plt.figure(figsize=(9,9))
example_images = X[:100]
plot_digits(example_images, images_per_row=10)
plt.show();
```

Let's split the data into our test and training sets before we inspect the data any further. The MNIST dataset is already split into a training set (the first 60,000 images) and a test set (the last 10,000 images).

```python
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

The training set is already shuffled for us which will help us for cross-validation scores to be consistent.

As a toy example, we will simplify the problem and only try to identify one digit - for example, the number 5. This will be a binary classifier that is capable of predicting just two classes - 5, and not 5.

```python
y_train_5 = (y_train == 5) # True for all 5s, False for everything else 
y_test_5 = (y_test == 5)
```

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(max_iter=1000, tol=0.001, random_state=42)
sgd_clf.fit(X_train, y_train_5)
```

We've chosen to use the SGDClassifier (Stochastic Gradient Descent Classifier) from scikit-learn to classifiy our 5s. We've instantiated the class, and fit our data to the model and now we will predict the class of the previously defined variable `some_digit`.

```python
sgd_clf.predict([some_digit])
```

The classifier predicted that the image stored in `some_digit` represents a 5 (True) which is the correct classification. Wwe'll now evaluate the model's performance.

For classification models, it's significantly harder to evaluate them as compared to regression models. We'll spend a great deal combing through the main ways to measure the performance of our model. We'll first dive into cross-validation to evaluate our model.

Let's borrow the `cross_val_score()` funcion from scikit-learn to evaluate the SGDClassifier model using K-fold cross-validation with three folds. K-fold cross-validation means we splitthe training set into K-folds, then make predictions and evaluate them on each fold using a model trained on the remaining folds.

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, X_train, y_train_5, cv=3, scoring="accuracy")
```

Sometimes we'll need more control over the cross-validation process than what scikit-learn provides. In this case, we can implement cross-validation ourselves. The following code does pretty much teh same thing as scikit-learn's `cross_val_score()` function and prints the same result.

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone

skfolds = StratifiedKFold(n_splits=3, random_state=42)

for train_index, test_index in skfolds.split(X_train, y_train_5):
    clone_clf = clone(sgd_clf)
    X_train_folds = X_train[train_index]
    y_train_folds = y_train_5[train_index]
    X_test_fold = X_train[test_index]
    y_test_fold = y_train_5[test_index]
    
    clone_clf.fit(X_train_folds, y_train_folds)
    y_pred = clone_clf.predict(X_test_fold)
    n_correct = sum(y_pred == y_test_fold)
    print(n_correct / len(y_pred))
```

We end up getting the exact same result as Scikit-learn's function scores.

It seems as though we have a very accurate model, acheiving a score above 95%. Let's look at an example of a dumb classifier that classifies all non-5 digits.

```python
from sklearn.base import BaseEstimator

class Never5Classifier(BaseEstimator):
    def fit(slef, X, y=None):
        pass
    def predict(self, X):
        return np.zeros((len(X), 1), dtype=bool)
    
never_5_clf = Never5Classifier()
cross_val_score(never_5_clf, X_train, y_train_5, cv=3, scoring="accuracy")
```

We get an accuracy of 90%, but this is because roughly 10% of the images in the dataset are 5s. If we guess that an image is not 5 then we'll be right 90% of the time. This demonstrates that accuracy is not the preferred performance measure for classifiers, especially when we're dealing with skewed datasets.

A better way to evaluate the performance of a classifier is to look at the confusion matrix. A confusion matrix works by counting the number of true positives, true negatives, false positives, and false negatives found when classifying images in the dataset.

To compute the confusion matrix we first need to have a set of predictions that can be compared to the actual targets. We can use `cross_val_predict()` for this.

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix

y_train_pred = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3)

confusion_matrix(y_train_5, y_train_pred)
```

After making the predictions first, we can then feed the predictions into the confusion_matrix function and compare the predicted values to the actual targets. We see that 53,892 examples were correctly classified as non-5s (true negatives), 687 were wrongly classified as 5s (false positives), 1,891 images were incorrectly classified as non-5s (false negatives) and 3,530 were correctly classified as 5s (true positives). Below is an example of a perfect classifier.

```python
y_train_perfect_predictions = y_train_5
confusion_matrix(y_train_5, y_train_perfect_predictions)
```

Peeking into the confusion matrix, we can look at the precision and recall of the classifier.

**Precision** = TP / TP + FP (TP = true positive; FP = false positive)

**Recall** = TP / TP + FN (TP = true positive; FN = false negative)

```python
from sklearn.metrics import precision_score, recall_score

print(precision_score(y_train_5, y_train_pred))
print(recall_score(y_train_5, y_train_pred))
```

The model doesn't look as good as it did when we were just measuring by accuracy. When it predicts that an image is a 5, it's correct only 83% of the time, while it only detects 65% of the 5s in the dataset.

Precision and recall can be combined into a single metric called an F-1 score. The F-1 score is the harmonic mean of precision and recall. A regular mean treats all values equally, whereas a harmonic mean gives much more weight to low values. As a result, the classifier will only get get high F-1 score if both recall and precision are high.

```python
from sklearn.metrics import f1_score

f1_score(y_train_5, y_train_pred)
```

**Note:** increasing precision reduces recall, and vice versa. This is called the precision/recall tradeoff.

To understand this tradeoff we'll look at how the SGDClassifier makes its classification decisions. For each instance it computes a score based on a decision function, and if that score if greater than a certain threshold, it assigns the instance to the positive class, anything else is assigned to the negative class.

```python
y_scores = sgd_clf.decision_function([some_digit])
y_scores
```

```python
threshold = 0
y_some_digit_pred = (y_scores > threshold)
y_some_digit_pred
```

```python
# Let's raise the threshold and see how the classifier responds
threshold = 8000
y_some_digit_pred = (y_scores > threshold)
y_some_digit_pred
```

This confirms that raising the threshold decreases the recall. The image prepresents a 5 and the classifier detects it when the thrashold is 0, but it misses it when the threshold is increased to 8000.

```python
y_scores = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3, method="decision_function")

from sklearn.metrics import precision_recall_curve

precisions, recalls, thresholds = precision_recall_curve(y_train_5, y_scores)

def plot_precision_recall_vs_thresholds(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], "b--", label='Precision', linewidth=2)
    plt.plot(thresholds, recalls[:-1], 'g-', label='Recall', linewidth=2)
    plt.legend(loc="center right", fontsize=16)
    plt.xlabel("Threshold", fontsize=16)
    plt.grid(True)
    plt.axis([-50000, 50000, 0, 1])
    
recall_90_precision = recalls[np.argmax(precisions >= 0.90)]
threshold_90_precision = thresholds[np.argmax(precisions >= 0.90)]

plt.figure(figsize=(8, 4))
plot_precision_recall_vs_thresholds(precisions, recalls, thresholds)
plt.plot([threshold_90_precision, threshold_90_precision], [0., 0.9], "r:")
plt.plot([-50000, threshold_90_precision], [0.9, 0.9], "r:")
plt.plot([-50000, threshold_90_precision], [recall_90_precision, recall_90_precision], "r:")
plt.plot([threshold_90_precision], [0.9], "ro")
plt.plot([threshold_90_precision], [recall_90_precision], "ro")
plt.show();
```

We may want to directly plot precision against recall to be able to select a good precision/recall tradeoff.

```python
def plot_precision_vs_recall(precisions, recalls):
    plt.plot(recalls, precisions, "b-", linewidth=2)
    plt.xlabel("Recall", fontsize=16)
    plt.ylabel("Precision", fontsize=16)
    plt.axis([0, 1, 0, 1])
    plt.grid(True)

plt.figure(figsize=(8, 6))
plot_precision_vs_recall(precisions, recalls)
plt.plot([0.4700, 0.4700], [0., 0.9], "r:")
plt.plot([0.0, 0.4700], [0.9, 0.9], "r:")
plt.plot([0.4700], [0.9], "ro")
plt.show();
```

Precision really starts to fall off around 80% so we will probably want to select a precision/recall tradeoff just before the drop off. 60% recall might be a good selection.

Let's pretend we want to aim for 90% precision. We can search for the lowest threshold that gives us at least 90% precision (np.argmax() will give us the first index of the maximum value, which in this case means the first True value):

```python
threshold_90_precision = thresholds[np.argmax(precisions >= 0.90)]
threshold_90_precision
```

To make predictions, instead of calling `predict()` method, we can just run the code below:

```python
y_train_pred_90 = (y_scores >= threshold_90_precision)

print(precision_score(y_train_5, y_train_pred_90))
print(recall_score(y_train_5, y_train_pred_90))
```

### ROC Curves
The receiver operator characteristic (ROC) curve is another common tool used with binary classifiers. It's very similar to precision/recall but instead of plotting precision versus recall, it calculates the true positive rate (another name for recall) against the false positive rate. A false positive rate is the ratio of negaitve instances that are incorrectly classified as positive. It's equal to 1 - true negative rate. The true negative rate is also called specificity. The ROC curve plots sensitivity (recall).

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_train_5, y_scores)
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--') # this is a dashed diagonal line
    plt.axis([0, 1, 0, 1])
    plt.xlabel("False Positive Rate (Fall-Out)", fontsize=16)
    plt.ylabel("True Positive Rate (Recall)", fontsize=16)
    plt.grid(True)
    
plt.figure(figsize=(8, 6))
plot_roc_curve(fpr, tpr)
plt.plot([4.837e-3, 4.837e-3], [0., 0.4368], "r:")
plt.plot([0.0, 4.837e-3], [0.4368, 0.4368], "r:")
plt.plot([4.837e-3], [0.4368], "ro")
plt.show();
```

Here we can see again that there is a tradeoff. The higher the true positive rate, more false positives are produced by the classifier. We can compare classifiers by measuring the area under the curve (AUC). A perfect classifier will have an ROC AUC of 1, and a random classifier will have an AUC ROC of 0.5. Scikit-learn provides a function `roc_auc_score()` to easily measure the area under the curve.

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_train_5, y_scores)
```

We would prefer to use the Precision/Recall curve as opposed to the ROC AUC score/ROC curve when the positive class is rare or when we care more about the false positive rate than the false negative rate. We can use the ROC curve any other time.

We'll train a RandomForestClassifier and compare it's ROC curve and ROC/AUC score to the SGDClassifier.

```python
from sklearn.ensemble import RandomForestClassifier

forest_clf = RandomForestClassifier(random_state=42)
y_probas_forest = cross_val_predict(forest_clf, X_train, y_train_5, cv=3,
                                   method="predict_proba")
```

To plot an ROC curve we need scores not probabilies. We'll use the positive class's probability as the score.

```python
y_scores_forest = y_probas_forest[:, 1]
fpr_forest, tpr_forest, thresholds_forest = roc_curve(y_train_5, y_scores_forest)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, "b:", linewidth=2, label="SGD")
plot_roc_curve(fpr_forest, tpr_forest, "Random Forest")
plt.plot([4.837e-3, 4.837e-3], [0., 0.4368], "r:")
plt.plot([0.0, 4.837e-3], [0.4368, 0.4368], "r:")
plt.plot([4.837e-3], [0.4368], "ro")
plt.plot([4.837e-3, 4.837e-3], [0., 0.9487], "r:")
plt.plot([4.837e-3], [0.9487], "ro")
plt.grid(True)
plt.legend(loc="lower right", fontsize=16)
plt.show();
```

#### Binary Classification Recap
 - We selected a classifier and trained it on our dataset
   - Stochastic Gradient Descent was our choice of classifier to start
 - Once the model was trained and we were able to get some predictions on it, we decided to evaulate the model on different performance measures
   - Started off by measuring accuracy using cross-validation (`cross_val_score()` function with K-fold cv)
   - Sometimes measuring the accuracy isn't too reliable. A better way to evaulate the performance is to look at the confusion matrix
   - Using the confusion matrix approach we have to make predictions on the training set first using `cross_val_predict()` and defining our classifier in it, the training set along with the training labels and how many fold we'd like
   - After `cross_val_predict()` we can call our `confusion_matrix()` function and pass in our training labels and our predicted labels. The output will show us how many true positives, true negatives, false positives and false negatives our classifier has predicted
   - Breaking down the metrics of the confusion matrix, we can look at the precision (accuracy of positive predictions) of the classifier and the recall (the ratio of positive instances correctly detected)
   - These two metrics can be combined into one, called the F-1 score
   - An f-1 score is the harmonic mean of precision and recall
   - One thing to note is that there is a tradeoff when it comes to precision and recall. The higher the precision the lower the recall, and vice versa. We'll want to optimize one or the other, or both, but all depending on the situation and how we want our model to be represented
   - The ROC curve is another common tool used to measure the performance of binary classifiers
     - The ROC curve plots the true positive rate against the false positive rate
   - Along with the ROC curve, we can measure the area under the curve, known as the ROC AUC score. This is used to compare classifiers with each other
     - We can select and train two different classifiers and compare their ROC AUC score to measure which one is better at predicting the correct labels

### Multiclass Classification
Some algorithms (such as Random Forest classifiers or naive Bayes classifiers) are capable of handling multiple classes directly. Others (such as Support Vector Machine classifiers or Linear classifiers) are strictly binary classifiers. However, there are various strategies that we can use to perform multiclass classification using multiple binary classifiers.

Two different strategies for classifiers are OvO (one versus one) or OvA (one versus all) classification.

```python
from sklearn.multiclass import OneVsOneClassifier
ovo_clf = OneVsOneClassifier(SGDClassifier(random_state=42))
ovo_clf.fit(X_train, y_train)
ovo_clf.predict([some_digit])
```

Scikit-learn detects when you're trying to use a binary classification algorithm for a multiclass classification task, and it automatically runs OvA (except for SVM classifiers for which it uses OvO. We'll try it with SGDClassifier

```python
sgd_clf.fit(X_train, y_train)
sgd_clf.predict([some_digit])
```

Under the hood, scikit-learn actually trained 10 binary classifiers, got their decision scores for the image and selected the class with the highest score

We're going to assume that we have found an optimal model and we want to find ways to improve it. One way to do this is to analyze the types of errors it makes. We first look at the confusion matrix. We have to make predictions using the `cross_val_predict()` function, then call the `confusion_matrix()` function.

```python
y_train_pred = cross_val_predict(sgd_clf, X_train_scaled, y_train, cv=3)
conf_mx = confusion_matrix(y_train, y_train_pred)
conf_mx
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(sgd_clf, X_train, y_train, cmap=plt.cm.gray)
```

```python
plt.matshow(conf_mx, cmap=plt.cm.gray)
plt.show();
```

The confusion matrix looks fairly good since most images are on the main diagonal which means that they were classified correctly. The 5s look slightly darker than the other digits, which could mean that there are fewer images of 5s in the dataset or that the classifier does not perform as well on 5s as on other digits. We can verify this by dividing each value in the confusion matrix by the numbers of images in the corresponding class. then we can compare the error rates instead of absolute number of errors.

```python
row_sums = conf_mx.sum(axis=1, keepdims=True)
norm_conf_mx = conf_mx / row_sums
```

```python
np.fill_diagonal(norm_conf_mx, 0)
plt.matshow(norm_conf_mx, cmap=plt.cm.gray)
plt.show();
```

Here we see that a lot of images get missclassified as 8s. Actual images of 8 get classified correctly though which is a good sign. We can also see that 3s and 5s get mixed up quite a bit. Analyzing the confusion matrix can often give some insights on ways to improve the classifier.

Analyzing individual erros can also be a good way to gain some insight but it's more difficult and time consuming to perform. An example of analyzing individual erros is shown below.

```python
cl_a, cl_b = 3, 5
X_aa = X_train[(y_train == cl_a) & (y_train_pred == cl_a)]
X_ab = X_train[(y_train == cl_a) & (y_train_pred == cl_b)]
X_ba = X_train[(y_train == cl_b) & (y_train_pred == cl_a)]
X_bb = X_train[(y_train == cl_b) & (y_train_pred == cl_b)]

plt.figure(figsize=(8,8))
plt.subplot(221); plot_digits(X_aa[:25], images_per_row=5)
plt.subplot(222); plot_digits(X_ab[:25], images_per_row=5)
plt.subplot(223); plot_digits(X_ba[:25], images_per_row=5)
plt.subplot(224); plot_digits(X_bb[:25], images_per_row=5)
plt.show()
```
